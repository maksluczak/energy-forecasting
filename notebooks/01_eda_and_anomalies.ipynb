```python
import matplotlib.pyplot as plt
import seaborn as sns

from src.data_loader import load_clean_data

df = load_clean_data()
print(df.head())
```

```python
plt.figure(figsize = (15, 5))
plt.plot(df.loc['2007-01-01':'2007-01-14', 'Global_active_power'], linewidth = 1)
plt.title('Energy usage - first two weeks January 2007')
plt.xlabel('Date')
plt.ylabel('Mean hourly energy usage [kW]')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (15, 5))
sns.boxplot(x = df['Global_active_power'])
plt.title('Value distribution')
plt.show()
```
